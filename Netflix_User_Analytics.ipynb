```python
# ============================================================
# NETFLIX USER ANALYTICS ASSIGNMENT
# ============================================================

# Q1-Q5 : DATASET UNDERSTANDING

import pandas as pd
import numpy as np

# Load Dataset
df = pd.read_csv("/content/Dataset 2.csv")

print("="*60)
print("Q1. First Five Records")
print(df.head())

print("\n" + "="*60)
print("Q2. Number of Rows and Columns")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\n" + "="*60)
print("Q3. Column Names")
print(df.columns.tolist())

print("\n" + "="*60)
print("Q4. Numerical Features")
print(df.select_dtypes(include=np.number).columns.tolist())

print("\nCategorical Features")
print(df.select_dtypes(include='object').columns.tolist())

print("\n" + "="*60)
print("Q5. Missing Values")
print(df.isnull().sum())

# ============================================================
# Q6-Q10 : EXPLORATORY DATA ANALYSIS
# ============================================================

print("\n" + "="*60)
print("Q6. Average Age")
print(df["Age"].mean())

print("\n" + "="*60)
print("Q7. Average Watch Hours Per Week")
print(df["WatchHoursPerWeek"].mean())

print("\n" + "="*60)
print("Q8. Average Monthly Spending")
print(df["MonthlySpend"].mean())

print("\n" + "="*60)
print("Q9. Users in Each Subscription Category")
print(df["SubscriptionType"].value_counts())

print("\n" + "="*60)
print("Q10. Percentage of Users Who Renewed")

renew_percent = (
    (df["SubscriptionRenewed"] == "Yes").sum()
    / len(df)
) * 100

print(f"{renew_percent:.2f}%")

# ============================================================
# Q11 : CONVERT CATEGORICAL FEATURES TO NUMERICAL
# ============================================================

from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

label_encoders = {}

categorical_columns = [
    "Gender",
    "SubscriptionType",
    "FavoriteGenre",
    "SubscriptionRenewed"
]

for col in categorical_columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    label_encoders[col] = le

print("\n" + "="*60)
print("Q11. Encoded Dataset")
print(df_encoded.head())

# ============================================================
# Q12 : DEFINE X AND y
# ============================================================

X = df_encoded.drop("SubscriptionRenewed", axis=1)
y = df_encoded["SubscriptionRenewed"]

print("\n" + "="*60)
print("Q12. Feature Set (X)")
print(X.head())

print("\nTarget Variable (y)")
print(y.head())

# ============================================================
# Q13 : TRAIN TEST SPLIT
# ============================================================

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("\n" + "="*60)
print("Q13. Train Test Split")
print("Training Size:", X_train.shape)
print("Testing Size:", X_test.shape)

# ============================================================
# Q14 : DECISION TREE MODEL
# ============================================================

from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train, y_train)

print("\n" + "="*60)
print("Q14. Decision Tree Model Trained Successfully")

# ============================================================
# Q15 : DECISION TREE ACCURACY
# ============================================================

from sklearn.metrics import accuracy_score

dt_pred = dt_model.predict(X_test)

dt_accuracy = accuracy_score(y_test, dt_pred)

print("\n" + "="*60)
print("Q15. Decision Tree Accuracy")
print(f"{dt_accuracy*100:.2f}%")

# ============================================================
# Q16 : CONFUSION MATRIX
# ============================================================

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, dt_pred)

print("\n" + "="*60)
print("Q16. Confusion Matrix")
print(cm)

# ============================================================
# Q17 : KNN CLASSIFIER (K = 5)
# ============================================================

from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(X_train, y_train)

print("\n" + "="*60)
print("Q17. KNN Model Trained Successfully")

# ============================================================
# Q18 : KNN ACCURACY COMPARISON
# ============================================================

knn_pred = knn_model.predict(X_test)

knn_accuracy = accuracy_score(y_test, knn_pred)

print("\n" + "="*60)
print("Q18. Accuracy Comparison")

print(f"Decision Tree Accuracy : {dt_accuracy*100:.2f}%")
print(f"KNN Accuracy           : {knn_accuracy*100:.2f}%")

if knn_accuracy > dt_accuracy:
    print("KNN performs better.")
elif dt_accuracy > knn_accuracy:
    print("Decision Tree performs better.")
else:
    print("Both models perform equally.")


# Q19 : LINEAR REGRESSION FOR MONTHLY SPENDING

from sklearn.linear_model import LinearRegression

X_reg = df_encoded.drop("MonthlySpend", axis=1)
y_reg = df_encoded["MonthlySpend"]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg,
    y_reg,
    test_size=0.20,
    random_state=42
)

lr_model = LinearRegression()

lr_model.fit(X_train_reg, y_train_reg)

print("\n" + "="*60)
print("Q19. Linear Regression Model Trained")


# Q20 : PREDICT MONTHLY SPENDING


sample_user = X_test_reg.iloc[[0]]

predicted_spending = lr_model.predict(sample_user)

print("\n" + "="*60)
print("Q20. Predicted Monthly Spending")
print(f"₹{predicted_spending[0]:.2f}")


# BUSINESS REFLECTION QUESTIONS


print("\n" + "="*60)
print("BUSINESS REFLECTION QUESTIONS")

print("\n1. Factors influencing subscription renewal:")
print("- Subscription Type")
print("- Monthly Spend")
print("- Watch Hours Per Week")
print("- Devices Used")
print("- Ad Clicks")

print("\n2. Why is subscription renewal a classification problem?")
print("Because the output is categorical (Yes/No).")

print("\n3. Why is monthly spending a regression problem?")
print("Because MonthlySpend is a continuous numerical value.")

print("\n4. Which algorithm performed better?")
if knn_accuracy > dt_accuracy:
    print("KNN performed better.")
elif dt_accuracy > knn_accuracy:
    print("Decision Tree performed better.")
else:
    print("Both performed equally.")

print("\n5. How can Netflix use these predictions?")
print("- Identify customers likely to cancel.")
print("- Provide personalized recommendations.")
print("- Offer targeted discounts.")
print("- Improve retention and revenue.")
```
